```python
# 爬取觀光局中的景點

import requests
from bs4 import BeautifulSoup
import os
import re

main_url = "https://www.taiwan.net.tw/"  # 觀光局首頁

city_url = "https://www.taiwan.net.tw/m1.aspx?sNo=0000505"  # 地區(人工複製網址 : 這次為"離島地區"，要改下面兩行網址)
city_response = requests.get(city_url)
city_soup = BeautifulSoup(city_response.content, "html.parser")

City_set = city_soup.find_all("a", class_="circularbtn")

# 取所有該縣市的區/鄉鎮市
for city in City_set:

  city_url = city.get("href")
  city_name = city.get("title")

  print("%s 有以下景點 : " % city_name)

  response = requests.get(main_url + city_url)
  soup = BeautifulSoup(response.content, "html.parser")

  # 解析推薦景點
  spots = soup.find_all("a", class_="card-link")

  # 把觀光局中推薦景點名稱分別地址正確的存成空資料夾

  for item in spots:

    # 取得景點名稱
    name = item.get("title")
    print("景點名稱 : %s" % name)

    temp_url = main_url + "/" + item.get("href")
    print("點擊後的網址為 : %s" % temp_url)
    temp_response = requests.get(temp_url)
    temp_soup = BeautifulSoup(temp_response.content, "html.parser")

    comp_address = temp_soup.find("a", class_="tel-link address").text
    print("該景點完整地址為 : %s" % comp_address)
    # 取得縣市名稱
    for i in os.listdir("/content/drive/MyDrive/旅遊/台灣/離島/%s" % city_name):

      if i in comp_address:
        try:
          townshipcity = re.search("%s(.+)"%i, comp_address).group(1)
          file_path = "/content/drive/MyDrive/旅遊/台灣/離島/%s/" % city_name + i + "/" + townshipcity + name
    
          if not os.path.isdir(file_path):
            os.makedirs(file_path)
          
          # 在文建夾中創立文字檔存放更多資訊
          file_path_info = file_path + "/" + "更多資訊.txt"
          info = open(file_path_info, "w")
          info.write("觀光部的資訊連結 : %s\n" % temp_url)
          info.close()
          break
          
        except:
          break
```
